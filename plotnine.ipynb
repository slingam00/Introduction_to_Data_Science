# plotnine

```python
from plotnine import *
import pandas as pd
```

```python
## Import iris data frame as example
data = pd.read_csv("iris.csv")
```

## Scatter Plot

```python
(ggplot(data, aes(x = "sepal_length", y = "sepal_width")) +
 geom_point()
)
```

## Differentiating the Scatter Plot by Species

```python
(ggplot(data, aes(x = "sepal_length", y = "sepal_width", fill = "species")) +   # filled by species - auto groups by the column
 geom_point(size = 3) # increased size
)
```
