# Matplotlib

https://matplotlib.org/tutorials/introductory/pyplot.html

```python
# Import Statements
import matplotlib.pyplot as plt
```

## Basic Plot

```python
plt.plot([1, 2, 3, 4])
plt.show()
```

## Labels for matplotlib

```python
plt.plot([1, 2, 3, 4])
plt.xlabel("Numbers")
plt.ylabel("Some More Numbers")
plt.title("Line Graph of Numbers")
plt.show()
```

## Graphing x- and y-coordinates

```python
plt.plot([0, 1, 2, 3, 4, 5], [0, 5, 12, 30, 60, 100])
plt.xlabel("X-Coordinates")
plt.ylabel("Y-Coordinates")
plt.title("Line Graph of X- and Y- Coordinates")
plt.show()
```

## Scatter Plot

```python
plt.scatter([1, 2, 3, 4, 5], [5, 12, 30, 60, 100])          

plt.xlabel("X-Coordinates")
plt.ylabel("Y-Coordinates")
plt.title("Scatter Plot Graph of X- and Y- Coordinates")
plt.show()
```

## Bar Graph

```python
fruits = ["apples", "oranges", "bananas"]
bars = [10, 6, 8]
plt.bar(fruits, bars)

plt.ylabel("Rating")
plt.xlabel("Fruits")
plt.title("Bar Graph showing Ratings of Fruit")
```
