# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4])
plt.show()
```

## Labels for matplotlib

```python
plt.plot([1, 2, 3, 4])
plt.xlabel("Numbers")
plt.ylabel("Some more Numbers")
plt.title("Line Graph of Numbers")
plt.show()
```

## Graphing x- and y-coordinates

```python
plt.plot([1, 2, 3, 4, 5], [5, 12, 30, 60, 100])
plt.show()
```

## Changing shape and color

```python
plt.plot([1, 2, 3, 4, 5], [5, 12, 30, 60, 100], 'go')
plt.show()
```
